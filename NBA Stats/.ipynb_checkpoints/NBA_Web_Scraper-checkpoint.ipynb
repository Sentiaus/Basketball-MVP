```python
years = list(range(1994,2023))
```

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
import requests
import os
```

```python
for year in years:
    url = url_start.format(year)
    data = requests.get(url)
    with open("mvp/{}.html".format(year), "w+",encoding="utf-8") as f: #Need encoding
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import io
from io import StringIO
```

```python
dfs = []

for year in years:
    with open("mvp/{}.html".format(year),encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page,"html.parser")
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find(id="mvp")
    mvp_table_html = str(mvp_table)
    mvp_table_io = StringIO(mvp_table_html)
    mvp = pd.read_html(mvp_table_io)[0]
    mvp["Year"] = year
    dfs.append(mvp)
```

```python
print(dfs[0])
```

```python
mvps = pd.concat(dfs)
```

```python
mvps.head()
```

```python
mvps.to_csv("mvps.csv")
```

```python
from selenium import webdriver
options = webdriver.EdgeOptions()
driver = webdriver.Edge(options=options)
import time
```

```python
player_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"
for year in years:
    url = player_stats_url.format(year)
    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)
    html = driver.page_source
    with open("player/{}.html".format(year), "w+", encoding = "utf-8") as f:
        f.write(html)
```

```python
df = []
for year in years:
    with open("player/{}.html".format(year), encoding = "utf-8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page,"html.parser")
    soup.find('tr', class_="thead").decompose()
    player_table = soup.find(id="per_game_stats")
    player_table_html = str(player_table)
    player_table_io = StringIO(player_table_html)
    player = pd.read_html(player_table_io)[0]
    player["Year"] = year
    df.append(player)
```

```python
players = pd.concat(df)
```

```python
players.to_csv("players.csv")
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
for year in years:
    url = team_stats_url.format(year)
    
    data = requests.get(url)
    
    with open("team/{}.html".format(year), "w+",encoding = "utf-8") as f:
        f.write(data.text)
```

```python
df = []
for year in years:
    with open("team/{}.html".format(year), encoding = "utf-8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page,"html.parser")
    soup.find('tr', class_="thead").decompose()
    team_table = soup.find(id="divs_standings_E")
    team_table_html = str(team_table)
    team_table_io = StringIO(team_table_html)
    team = pd.read_html(team_table_io)[0]
    team["Year"] = year
    team["Team"] = team["Eastern Conference"]
    del team["Eastern Conference"]
    df.append(team)
    
    soup = BeautifulSoup(page,"html.parser")
    soup.find('tr', class_="thead").decompose()
    team_table = soup.find(id="divs_standings_W")
    team_table_html = str(team_table)
    team_table_io = StringIO(team_table_html)
    team = pd.read_html(team_table_io)[0]
    team["Year"] = year
    team["Team"] = team["Western Conference"]
    del team["Western Conference"]
    df.append(team)
```

```python
teams = pd.concat(df)
teams
```

```python
teams.to_csv("teams.csv")
```
